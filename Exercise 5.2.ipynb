```python
from operator import itemgetter
filename = "D:\\Development\\pythonworkout\\temp\\passwd.txt"

users = {}
with open(filename) as f:
    for line in f:
        if not line.startswith('#') and line.strip():
            user_info = line.split(':')
            users[user_info[0]] = int(user_info[2])

for username, user_id in sorted(users.items(), key=itemgetter(1)):
    print(f"{username:25}:{user_id}")
```
