```python
import json
import pathlib

scores_dir = "D:\\Development\\pythonworkout\\temp\\scores"
p = pathlib.Path(scores_dir)

def gather_scores(students:list[dict]) -> dict:
    results = {}
    for student in students:
        for subject, score in student.items():
            results.setdefault(subject, [])
            results[subject].append(score)
    return results

def compute_statistics(scores: dict) -> dict:
    return {k:{'min':min(v), 'max':max(v), 'average':round(sum(v)/len(v),1)} for k,v in scores.items()}

for file in p.iterdir():
    subjects = set()
    with open(file) as f:
        stats_by_subject = compute_statistics(gather_scores(json.load(f)))
        print(f"Class {file.name}")
        for subject, stats in stats_by_subject.items():
            print(f"\t{subject}: {", ".join(["" + k + " " + str(v) +"" for k,v in stats.items()])}")
```
